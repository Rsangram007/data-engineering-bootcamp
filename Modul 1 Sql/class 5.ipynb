```sql
--Gives each row a unique number based on the order_date.

select
order_id,
order_date,
row_number() over(order by order_date) as rownum
from order
```

```sql
-- Real-Life Use Case: Show first product bought by each customer

select
customer.first_name,
orderdetails.product_name,
orderdetails.order_id,

row_number() over (partition by customer.customer_id order by order.order_date) as purchessequency
from 
customer 
join order on customer.customer_id = order.customer_id
join orderdetails on order.order_id = orderdetails.order_id
```

```sql
---The First order is that where purchase_sequence=1
--RANK()
--Ranks rows by quantity; ties get same rank, but gaps are left in the sequence.

select 
 
quantity,
product_name,

rank() over (order by quantity desc) as ranking
from orderdetails
```

```sql
-- Real-Life Use Case: Assign bonus based on best-selling products in each order

select
order_id,
product_name,
quantity,

rank() over (partition by order_id order by quantity desc) as ranking
from 
orderdetails
```

```sql
select
  product_name,
  quantity,
  dense_rank() over(order by quantity desc) as denserank,
  rank() over (order by quantity desc) as rank
from
  orderdetails
```

```sql
--  Real-Life Use Case: Label product popularity tiers in clean rank buckets

select
product_name,
sum(quantity) as totalqty,

dense_rank() over( order by sum(quantity) desc) as popularityrank
from orderdetails
group by 1
```

```sql
--LAG() - COMPARE WITH PREVIOUS ROWS

select 
order_id,order_date,
lag(order_date) over (order by order_date) as priviousdate
from order
```

```sql
--LEAD() - COMPARE WITH NEXT ROWS

select
order_id,order_date,
lead(order_date) over(order by order_date) as nextorderdate
from order
```

```sql
--Real-Life Use Case: Predict what customers might buy next

select
c.first_name,
od.product_name,

lead(od.product_name) over (partition by c.customer_id order by o.order_date ) as possiablenextproduct
from customer as c

join order as o on c.customer_id = o.customer_id

join orderdetails as od on o.order_id = od.order_id
```

```sql
--SUM() ---> RUNNING TOTAL OR Cummulative Sum 

select 
order_id,product_name,quantity,
sum (quantity) over (order by order_id) as cumsum
from orderdetails
```
