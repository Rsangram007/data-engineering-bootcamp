```sql
-- Find customers who placed orders above the average order amount.

select
c.first_name,
c.last_name,
o.total_amount

from customer as c inner join order as o on c.customer_id=o.customer_id where o.total_amount >

(select avg(total_amount) from order)

```

```sql

-- List all products that have been sold more than 5 times in total.

select
od.product_name ,sum(od.quantity)
from orderdetails as od  group by od.product_name having sum(od.quantity) >

(select 2)
```

```sql
--  List customers and their total quantity of products ordered.

with customerorder as (
select 
upper( concat( c.first_name , " ", c.last_name)  ) as customername ,sum(od.quantity) as totalquanity,c.customer_id

from customer as c
inner join order as o on c.customer_id=o.customer_id
inner join orderdetails as od on o.order_id=od.order_id
group by c.first_name ,o.order_id,od.product_name,c.last_name,c.customer_id

)


select * from customerorder where totalquanity>3
```

```sql
--Find products with their total sales revenue (price * quantity).

with totalrevenue as (

  select
 od.product_id, od.product_name as Productname,od.price_each,od.quantity,sum(od.price_each * od.quantity)as revenue 
  from orderdetails as  od
  group by product_name,price_each,quantity,product_id
)


select * from totalrevenue where revenue >7000
```

```sql
--Find products with their total sales revenue (price * quantity).with out groupby

with totalrevenue as (

  select
   sum(od.price_each * od.quantity)as revenue  
  from orderdetails as  od
  
)


select * from totalrevenue where revenue >7000
```

```sql
--Categorize each order as 'Small', 'Medium', 'Large'.

select order.order_id,order.total_amount,
case 
when order.total_amount < 500 then "small"
when order.total_amount between 500 and 1000 then "medium"
when order.total_amount > 1000 then "large"
end 

from order
```

--Label customers as 'New' or 'Returning' based on order count.

```sql
select c.first_name,c.last_name,c.customer_id, count(o.order_id) as totalorder,

case 

when count(o.order_id)=1 then "new"
when count(o.order_id)>1 then "returning"
else "No return"


end as returndetails

from customer as c 
left join order as o on c.customer_id=o.customer_id
group by c.first_name,c.last_name,c.customer_id
```

```sql
SELECT c.first_name, COUNT(o.order_id) AS total_orders,
       CASE
           WHEN COUNT(o.order_id) = 1 THEN 'New'
           WHEN COUNT(o.order_id) > 1 THEN 'Returning'
           ELSE 'No Orders'
       END AS customer_type
FROM customer c
LEFT JOIN order o ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.first_name;
```
