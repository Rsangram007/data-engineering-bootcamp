```sql
 ---Badly optimisated query
SELECT * 
FROM customer c
JOIN order o ON c.customer_id = o.customer_id
JOIN orderdetails od ON o.order_id = od.order_id;

```

```sql
--OPTIMISATED QUERY 

SELECT
c.first_name,
c.last_name,
od.product_name,
od.price_each,
o.order_id
from customer as c
 join order as o  on c.customer_id=o.customer_id
 join orderdetails as od on o.order_id=od.order_id
```

```sql

--âŒ Unoptimized:

SELECT c.first_name, od.product_name, od.price_each
FROM customer c
JOIN order o ON c.customer_id = o.customer_id
JOIN orderdetails od ON o.order_id = od.order_id
ORDER BY od.price_each DESC;
```

```sql
-- âœ… Optimized:

SELECT c.customer_id, SUM(od.price_each * od.quantity) AS total_spent
FROM customer c
JOIN order o ON c.customer_id = o.customer_id
JOIN orderdetails od ON o.order_id = od.order_id
WHERE o.order_date >= '2023-01-01'
GROUP BY c.customer_id;
```

```sql
EXPLAIN
SELECT c.first_name, o.order_id, od.product_name
FROM customer c
JOIN order o ON c.customer_id = o.customer_id
JOIN orderdetails od ON o.order_id = od.order_id
WHERE od.price_each > 1000;
```
